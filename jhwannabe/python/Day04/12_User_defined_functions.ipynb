# 1. 함수
* 변수 x, y에 대해 x값이 정해지면 그에 따라 y값을 결정하는 것
* 반복적으로 사용되는 가치있는 부분을 묶어서 하나의 집합으로 만듦
* 코드를 재활용하기 위한 목적

## 1-1. 이름만 있는 함수
```
def 함수명():
    함수가 호출되면 실행할 문장
```
```
함수명() # 호출
```

````python
def func1():
    print('처음으로 만드는 함수!')
````

````python
func1()
````

````python
for i in range(5):
    func1()
````

## 1-2. 매개변수가 있는 함수
```
def 함수명(변수1, 변수2, ...)
    함수가 호출되면 실행할 문장
```

```
함수명(값1, 값2, ...) # 호출
```

````python
def func2(num):
    print(f'입력받은 숫자는 {num}')
````

````python
func2(4)
````

````python
def func3(start, end):
    sum = 0
    for i in range(start, end+1):
        sum += i
    print(f'{start}부터 {end}까지의 합은 {sum}')
````

````python
func3(5,23)
````
