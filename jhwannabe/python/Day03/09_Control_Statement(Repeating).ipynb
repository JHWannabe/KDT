## 1-2. 대입 연산자
* =, += -=, *=, /=, %=
* 왼쪽에 저장된 값과 오른쪽에 값을 연산한 뒤 다시 왼쪽에 대입(inplace 연산) 

````python
num = 10
````

````python
num += 1
print(num)
num -= 1
print(num)
num *= 10
print(num)
num /= 10
print(num)
num %= 2
print(num)
````

## 1-3. While 문
```
while 조건식:
    조건식의 결과가 True인 동안 반복할 문장
```

````python
# 무한 루프
# while True:
#  print('끝나지 않는 반복')
````

````python
i = 1

while i <= 5:
  print('Hello Python!')
  i += 1

print("현재 i의 값:", i)
````

````python
# 1 ~ 100까지의 총합
i = 1
sum = 0

while i <= 100:
  sum += i
  i += 1

print('총합:', sum)
````

## 문제 1
* 1 ~ 100까지의 짝수의 총합을 구하라

````python
i = 1
sum = 0

while i <= 100:
  if i % 2 == 0:
    sum += i
  i += 1

print('총합:', sum)
````

## 1-5. enumerate
* 반복문 사용 시 몇번째 반복인지 인덱스 번호로 확인
* 인덱스 번호와 요소를 튜플 형태로 변환

````python
li1 = [10, 20, 30, 40, 50]

for e in enumerate(li1):
  print(e)
````

````python
for i, v in enumerate(li1):
  print(f'index: {i}, value: {v}')
````

## 1-6. zip() 함수
* 객체가 담고 있는 원소를 엮어서 튜플의 형태로 반환
* 매개변수의 길이가 다를 때는 짧은 매개변수 기준으로 데이터가 엮이고, 나머지는 버려짐

````python
li1 = [10, 20, 30]
li2 = ['apple','banana','orange']
# (10, 'apple')
# (20, 'banana')
# (30, 'orange')

# 좋지 않은 코드
for i in range(len(li1)):
  print((li1[i], li2[i]))
````

````python
for l1, l2 in zip(li1, li2):
  print((l1, l2))
````

````python
for n, A, a in zip('12345', 'ABCDE', 'abcde'):
  print((n, A, a))
````

````python
for n, A, a in zip('123456', 'ABCDE', 'abcde'):
  print((n, A, a))
````

## 문제 4
* 아래 score 리스트에 저장된 점수가 60점 이상인 학생이 몇 명인지 알아보는 프로그램을 작성해보자

````python
score = [90, 30, 50, 60, 80, 70, 100, 40, 20, 10]
````

````python
count = 0;

for i in score:
  if i >= 60:
    count += 1

print('60점 이상인 학생 수:', count)
````

# 2. 다중 반복문
* 반복문이 2개이상 겹쳐져 있는 형태

````python
for i in range(1, 4):
  print(f'😎 i: {i}')
  for j in range(1, 4):
    print(f'   😍 i: {j}')
````

## 문제 3
* 2단부터 9단까지 구구단을 출력하는 프로그램을 while문으로 작성해보자

````python
i = 2

while i < 10:
  print(f'{i}단')
  j = 1
  while j < 10:
    print(f'{i} x {j} = {i*j}')
    j += 1
  i += 1
  print()
````
