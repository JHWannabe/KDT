## 1-1. 자료형 참/거짓 판단하기
* 파이썬에서 사용하는 모든 자료형은 참 또는 거짓으로 표현이 가능

````python
bool1 = True
print(bool1)
print(type(bool1))
````

````python
# 0이 아닌 숫자는 모두 True, 0은 False

num1 = 10
print(num1)
print(type(num1))
print(bool(num1)) # 모든 값들은 bool형으로 형변환 가능 True

num2 = 0
print(num2)
print(type(num2))
print(bool(num2)) # False
````

````python
str1 = '안녕'
print(str1)
print(type(str1))
print(bool(str1)) # True

str2 = ''
print(str2)
print(type(str2))
print(bool(str2)) # False
````

````python
list1 = [1, 2, 3]
print(list1)
print(type(list1))
print(bool(list1)) # True

list2 = []
print(list2)
print(type(list2))
print(bool(list2)) # False
````

````python
list1 = [1, 3, 5]

if list1:
  print('요소가 있습니다.')
else:
  print('요소가 없습니다.')
````

## 1-2. 대입 연산자
* =, += -=, *=, /=, %=
* 왼쪽에 저장된 값과 오른쪽에 값을 연산한 뒤 다시 왼쪽에 대입(inplace 연산) 

````python
num = 10
````

````python
num += 1
print(num)
num -= 1
print(num)
num *= 10
print(num)
num /= 10
print(num)
num %= 2
print(num)
````

## 1-3. While 문
```
while 조건식:
    조건식의 결과가 True인 동안 반복할 문장
```

````python
# 무한 루프
while True:
 print('끝나지 않는 반복')
````

````python
i = 1

while i <= 5:
  print('Hello Python!')
  i += 1

print("현재 i의 값:", i)
````

````python
# 1 ~ 100까지의 총합
i = 1
sum = 0

while i <= 100:
  sum += i
  i += 1

print('총합:', sum)
````

## 문제 1
* 1 ~ 100까지의 짝수의 총합을 구하라

````python
i = 1
sum = 0

while i <= 100:
  if i % 2 == 0:
    sum += i
  i += 1

print('총합:', sum)
````

## 문제 2
* 원하는 단을 입력받아 해당 단의 구구단을 출력하라

````python
num = int(input('원하는 단을 입력하세요:'))
i = 1

print('%d단' % num)
while i < 10:
  print('%d * %d = %d' % (num,i,num*i))
  i += 1
````
