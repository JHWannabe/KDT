```python
class Movie:
    def __init__(self, name, genre, time, theater):
        
        self.mv_name = name
        self.mv_genre = genre
        self.mv_time = time
        self.mv_theater = theater

    # setter
    # def setMv_num(self, num):
    #     self.mv_num = num

    def setMv_name(self, name):
        self.mv_name = name

    def setMv_genre(self, genre):
        self.mv_genre = genre

    def setMv_time(self, time):
        self.time = time

    def setMv_theater(self, theater):
        self.theater = theater

    # getter
    # def getMv_num(self):
    #     return self.mv_num

    def getMv_name(self):
        return self.mv_name

    def getMv_genre(self):
        return self.mv_genre

    def getMv_time(self):
        return self.mv_time

    def getMv_theater(self):
        return self.mv_theater
```

```python
import MySQLdb

class movieList:
    def __init__(self):
        self.db = None

    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'hw')
        
    def disconnect(self):
        self.db.close()

    def insert(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = "insert into movie values (%s, %s, %s, %s)"
        data = (movie.getMv_name(), movie.getMv_genre(), movie.getMv_time(), movie.getMv_theater())
        cur.execute(sql, data)
        self.db.commit()
        self.disconnect() 

    def update(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = 'update movie set mv_genre=%s, mv_time=%s, mv_theater=%s where mv_name=%s'
        data = (movie.getMv_genre(), movie.getMv_time(), movie.getMv_theater(), movie.getMv_name())
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 단어가 없습니다')
        self.disconnect()

    def search(self, mv_name):

        self.connect()
        cur = self.db.cursor()
        sql = 'select * from movie where mv_name=%s'
        data = (mv_name,)
        cur.execute(sql, data)
        row = cur.fetchone()
        self.disconnect()
        return row
    
    def selectAll(self):
        self.connect()
        cur = self.db.cursor()
        sql = 'select * from movie'
        cur.execute(sql)
        row = cur.fetchall()
        self.disconnect()
        return row

    def delete(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = 'delete from movie where mv_name=%s'
        data = (movie,)
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('삭제되었습니다')
        else:
            print('해당 영화 정보가 없습니다')
        self.disconnect()
```

```python
from datetime import datetime

now = datetime.now()

class Review:
    def __init__(self, name, score, contents):
        self.name = name
        self.score = 5
        self.contents = contents
        self.time = now.strftime('%Y-%m-%d %H:%M:%S')

    # setter
    def setName(self, name):
        self.name = name

    def setScore(self, score):
        self.score = score

    def setContents(self, contents):
        self.contents = contents

    # getter
    def getName(self):
        return self.name

    def getScore(self):
        return self.score

    def getContents(self):
        return self.contents
    
    # print
    def printReview(self):
        print(f'이름: {self.name}\n리뷰 내용: {self.contents}\n작성 시간: {self.time}\n\n')
```

```python
class reviewList:
    def __init__(self):
        self.datas= []

    def insert(self, review):
        self.datas.append(review)

    def update(self, review):
        for i in self.datas:
            if i.getName() == review.getName(): # 기존에 저장된 단어와 수정된 단어가 같다면
                i.setScore(review.getScore()) # 기존에 저장된 뜻에 수정할 단어의 뜻을 설정
                i.setContents(review.getContents()) # 기존에 저장된 레벨에 수정할 단어의 뜻을 설정
        

    def search(self, review):
        for i in self.datas:
            if i.getName() == review:
                return i
    
    def selectAll(self):
        return self.datas

    def delete(self, review):
        self.datas.remove(review)
```
