```python
class Movie:
    def __init__(self, name, genre, time, theater):
        self.mv_name = name
        self.mv_genre = genre
        self.mv_time = time
        self.mv_theater = theater

    # setter
    def setMv_name(self, name):
        self.mv_name = name

    def setMv_genre(self, genre):
        self.mv_genre = genre

    def setMv_time(self, time):
        self.time = time

    def setMv_theater(self, theater):
        self.theater = theater

    # getter
    def getMv_name(self):
        return self.mv_name

    def getMv_genre(self):
        return self.mv_genre

    def getMv_time(self):
        return self.mv_time

    def getMv_theater(self):
        return self.mv_theater
```

```python
import MySQLdb

class movieList:
    def __init__(self):
        self.db = None

    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'hw')

    def disconnect(self):
        self.db.close()

    def insert(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = "insert into movie (mv_name, mv_genre, mv_time, mv_theater) values (%s, %s, %s, %s)"
        data = (movie.getMv_name(), movie.getMv_genre(), movie.getMv_time(), movie.getMv_theater())
        cur.execute(sql, data)
        self.db.commit()
        self.disconnect()

    def update(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = 'update movie set mv_genre=%s, mv_time=%s, mv_theater=%s where mv_name=%s'
        data = (movie.getMv_genre(), movie.getMv_time(),
                movie.getMv_theater(), movie.getMv_name())
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 영화 정보가 없습니다')
        self.disconnect()

    def search(self, mv_name):
        self.connect()
        cur = self.db.cursor()
        sql = 'select * from movie where mv_name=%s'
        cur.execute(sql, (mv_name,))
        row = cur.fetchone()
        self.disconnect()
        return row
    
    def reserveSearch(self, mv_num):
        self.connect()
        cur = self.db.cursor()
        sql = 'select * from movie where mv_num=%s'
        cur.execute(sql, (mv_num,))
        row = cur.fetchone()
        self.disconnect()
        return row

    def printAll(self):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from movie'
        cur.execute(sql)
        while 1:
            movie = cur.fetchone()
            if movie:
                print(f"영화 번호: {movie['mv_num']}, 영화명: {movie['mv_name']}, 장르: {movie['mv_genre']}, 시간: {movie['mv_time']}, 상영관: {movie['mv_theater']}")
            else:
                break
        self.disconnect()
        return movie

    def delete(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = 'delete from movie where mv_name=%s'
        result = cur.execute(sql, (movie,))
        self.db.commit()
        if result > 0:
            print('삭제되었습니다')
        else:
            print('해당 영화 정보가 없습니다')
        self.disconnect()
        
class reserveList:
    def __init__(self):
        self.db = None

    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'hw')

    def disconnect(self):
        self.db.close()
        
    def update(self, movie):
        self.connect()
        cur = self.db.cursor()
        sql = 'update movie set mv_genre=%s, mv_time=%s, mv_theater=%s where mv_name=%s'
        data = (movie.getMv_genre(), movie.getMv_time(),
                movie.getMv_theater(), movie.getMv_name())
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 영화 정보가 없습니다')
        self.disconnect()

    def search(self, num):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from reservation where res_hp=%s'
        cur.execute(sql, (num,))
        while 1:
            rs = cur.fetchone()
            if rs:
                print(f"예매 번호: {rs['res_num']}, 영화명: {rs['res_mv_name']}, 상영시간: {rs['res_mv_time']}, 상영관: {rs['res_mv_theater']}, 좌석: {rs['res_seat']}")
            else:
                break
        self.disconnect()
        return rs
    
    def reserveSearch(self, mv_num):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from reservation where mv_num=%s'
        cur.execute(sql, (mv_num,))
        row = cur.fetchone()
        self.disconnect()
        return row

    def printAll(self):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from reservation'
        cur.execute(sql)
        while 1:
            rs = cur.fetchone()
            if rs:
                print(f"예매 번호: {rs['res_num']}, 영화명: {rs['res_mv_name']}, 상영시간: {rs['res_mv_time']}, 상영관: {rs['res_mv_theater']}, 좌석: {rs['res_seat']}")
            else:
                break
        self.disconnect()
        return rs

#     def delete(self, movie):
#         self.connect()
#         cur = self.db.cursor()
#         sql = 'delete from movie where mv_name=%s'
#         result = cur.execute(sql, (movie,))
#         self.db.commit()
#         if result > 0:
#             print('삭제되었습니다')
#         else:
#             print('해당 영화 정보가 없습니다')
#         self.disconnect()
```

```python
from datetime import datetime

class manageMode:
    def __init__(self):
        self.movielist = movieList()

    def addMovie(self):
        name = input('영화명을 입력하세요: ')
        genre = input('영화장르를 입력하세요: ')
        time = input('상영시간을 입력하세요: ')
        theater = input('상영관를 입력하세요: ')

        movie = Movie(name, genre, time, theater)
        self.movielist.insert(movie)

    def delMovie(self):
        del_movie = input('삭제할 영화를 입력하세요: ')
        movie = self.movielist.search(del_movie)
        if movie == None:
            print('삭제할 영화가 없습니다.')
        else:
            self.movielist.delete(movie)

    def printAllMovie(self):
        datas = self.movielist.printAll()
        

    def searchMovie(self):
        search_name = input('검색할 영화를 입력하세요: ')
        movie = self.movielist.search(search_name)
        if movie:
            print(movie)
        else:
            print('찾는 영화가 없습니다')

    def updateMovie(self):
        name = input('수정할 영화를 입력하세요: ')
        movie = self.movielist.search(name)
        if movie == None:
            print('수정할 영화가 없습니다.')
        else:
            genre = input('새로운 장르를 입력하세요: ')
            time = input('새로운 상영시간을 입력하세요: ')
            theater = input('새로운 상영관을 입력하세요: ')
            movie = Movie(name, genre, time, theater)
            self.movielist.update(movie)

    def delMovie(self):
        name = input('삭제할 영화를 입력하세요: ')
        self.movielist.delete(name)


class customMode:
    def __init__(self):
        self.movielist = movieList()
        self.reservelist = reserveList()

    def selectMovie(self):
        hp = int(input('예매 확인 번호(전화번호 뒷 4자리)를 입력하세요: '))
        self.movielist.printAll()
        num = input('예매할 영화 번호를 입력하세요: ')
        mv = list(self.movielist.reserveSearch(num))
        people = int(input('예매 인원을 입력하세요(최대 3명, 숫자만 입력): '))

        for _ in range(0,people):
            print('좌석은 A~M, 1~10을 선택할 수 있습니다.')
            seat = input('예매할 좌석을 입력하세요 (E5): ')
            seat = seat.upper()
            time = datetime.now()
            time = time.strftime('%Y-%m-%d %H:%M:%S')

            self.reservelist.connect()
            cur = self.reservelist.db.cursor(MySQLdb.cursors.DictCursor)
            sql = "insert into reservation (res_hp, res_mv_name, res_mv_time, res_mv_theater, res_seat, res_time) values (%s, %s, %s, %s, %s, %s)"
            data = (hp, mv[1], mv[3], mv[4], seat, time)
            result = cur.execute(sql, data)
            self.reservelist.db.commit()

        print('예매가 완료되었습니다.')
            
    def printReserve(self):
        num = int(input('예매 확인 번호(전화번호 뒷 4자리)를 입력하세요: '))
        self.reservelist.search(num)
        
```
