# 2. 셀의 종류
* 코드: 파이썬 코드 및 리눅스 명령어를 실행할 수 있는 셀
* 텍스트: 설명을 달거나 문구 등을 적기 위한 텍스트 전용 셀(마크다운)

##2-1. 셀의 삽입 방법
1. 상단의 '+코드', '+텍스트' 버튼을 눌러 삽입
2. 코드셀을 위에 삽입: Ctrl + M, A
3. 코드셀을 밑에 삽입: Ctrl + M, B

##2-2. 셀의 변경 방법
1. 코드 셀로 변경: Ctrl + M, Y
2. 마크다운 셀로 변경: Ctrl + M, M


##2-3. 셀의 취소 방법
1. 셀 지우기: Ctrl + M, D
2. 실행 취소: Ctrl + M, Z

# 3. 셀의 실행
1. Ctrl + Enter: 해당 셀을 실행하고 커서는 해당 셀에 남음
2. Shift + Enter: 해당 셀을 실행하고 커서는 다음 셀로 이동
3. Alt + Enter: 해당 셀을 실행하고 커서는 다음 셀을 삽입 후 이동
