# 4. 폴더 생성하기

```python
import fnmatch
```

```python
def categoryList(target_path : str) -> list:
    file_list = []
    for filename in os.listdir(target_path):
        if fnmatch.fnmatch(filename, '*.mp3'):
            file_list.append(filename)
    
    category = []
    for file in file_list:
        temp_list = file.split('-')
        category.append(temp_list[0])
    
    temp_set = set(category)
    result = list(temp_set)
    return result
```

```python
categoryList(target_path)
```

```python
categorylist = categoryList(target_path) + ['기타']
print(len(categorylist))
```

```python
import pathlib

new_path = './new_dir'
def makeDir(new_path: str, categorylist: list):
    for category in categorylist:
        new_dir = pathlib.Path(os.path.join(new_path, category))
        new_dir.mkdir(parents=True, exist_ok = True) # 존재하면 덮어쓰기
```

```python
makeDir(new_path,categorylist)
```

# 5. 파일 분류 및 이동

```python
import shutil

def moveFile(new_path, target_path, categorylist):
    dirlist = os.listdir(new_path) # 이동시킬 경로에 생성된 분류 디렉토리 리스트
    filelist = os.listdir(target_path) # 이동시킬 파일명 리스트
    categorydic = {}
    
    for file in filelist:
        try:
            temp_list = file.split('-')
            assert temp_list[0] in categorylist # 카테고리가 맞지 않으면 에러 발생
            temp_list[0] = temp_list[0].strip()
            categorydic[file] = temp_list[0] # {'파일명': '분류명'}
        except:
            categorydic[file] = '기타'
    
#     print(categorydic)

    for key, value in categorydic.items():
        shutil.copy(target_path + '/' + key, new_path + '/' + value)
```

```python
moveFile(new_path, target_path, categorylist)
```

```python
# 폴더별 파일명을 입력받아 엑셀파일에 저장하는 함수
def getFileName(target_path, new_path):   
    wb = opx.Workbook()
    ws = wb.active # 새로 생성한 WorkBook의 활성화 시트를 ws로 정의
    
    ws.cell(row=1, column=1).value = '파일명'
    ws.cell(row=1, column=2).value = '파일 경로(변경 전)'
    ws.cell(row=1, column=3).value = '파일 경로(변경 후)'
    
    i = 2
    
    before_dir = target_path
    filelist = os.listdir(before_dir)
    after_dir = new_path
    
    for filename in filelist:
        artist = filename.split('-')
        ws.cell(row=i, column=1).value = filename
        ws.cell(row=i, column=2).value = before_dir + '/'   
        ws.cell(row=i, column=3).value = after_dir + '/' + artist[0]
        i = i+1
        
    wb.save('musiclist.xlsx')
```

```python
getFileName(target_path, new_path)
```
