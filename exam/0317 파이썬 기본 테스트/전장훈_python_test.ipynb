1. 이름, 전화번호, 주소, 이메일, 그룹을 입력받아 저장합니다. (10점) - 입력

2. 입력받은 이름, 전화번호, 주소, 이메일, 그룹을 출력합니다. (5점) - 출력

3. 이름, 전화번호 일부, 이메일, 그룹 등으로 전화번호를 검색할 수 있습니다. (10점) - 검색

4. 전화번호를 삭제할 수 있습니다. (5점) - 삭제

5. 저장된 전화번호를 파일에 저장합니다. (10점) - 저장

6. 프로그램 종료 후 재실행 할 때 저장된 전화번호를 불러옵니다. (10점)

7. 전화번호 출력시 이름순으로 오름차순 정렬합니다. (10점) - 이름 정렬

8. 중복된 전화번호는 저장할 수 없습니다. (10점)

9. 적절한 예외처리를 하여 비정상적인 프로그램 종료가 생기지 않습니다. (5점)

10. 정상 종료시까지 무한으로 메뉴를 반복합니다. (5점)

11. 각 라인에 대해 자세한 주석을 작성합니다. (10점)

12. 클래스를 이용하여 객체지향 프로그래밍으로 설게합니다. (10점)

 

* 각 문제의 부분 점수는 아래와 같습니다.

10점 문제

상(10점), 중(7점), 하(5), 에러(3)

5점 문제

상(5점), 중(3점), 하(2점), 에러(1)

```python
class PhoneInfo:
    def __init__(self):     # 생성자
        self.num = ''       # 메뉴 선택 숫자
        self.search = ''    # 검색 정보 저장 문자열

    # 정보 입력 함수
    def insertNum(self):
        print('정보를 입력하세요')
        name = input('이름: ')
        hp = input('전화번호(-포함): ')

        # 전화번호 중복 여부
        for i in range(len(phonelist)):
            if hp == phonelist[i][1]:
                print('중복된 정보가 있습니다.')
                return 0
        
        address = input('주소: ')
        email = input('이메일: ')
        group = input('그룹: ')

        # input으로 받은 모든 변수를 리스트로 변환
        result = [name,hp,address,email,group]
        # 메모리에 저장
        phonelist.append(result)
        print('정보가 입력되었습니다.')

    # 정보 출력 함수
    def printNum(self):
        if len(phonelist) == 0:
            print('리스트에 저장된 정보가 없습니다.')
        else:
            # 리스트 정렬
            phonelist.sort(key=lambda x:x[0])
            for i in range(len(phonelist)):
                print('이름: ' + phonelist[i][0] + ', 전화번호: ' + phonelist[i][1] + ', 주소: ' + phonelist[i][2] + ', 이메일: ' + phonelist[i][3] + ', 그룹: ' + phonelist[i][4])
    
    # 정보 저장 함수
    def saveNum(self):
        file = open('phonelist.txt', 'w')
        for i in range(len(phonelist)):
            file.write(str(phonelist[i]))
            if i != len(phonelist)-1:
                file.write("\n")
        file.close()
        print('저장되었습니다.')

    # 상세 메뉴 함수
    def inputNum(self,num):
        # 이름 검색
        if num == 1:
            search = input('이름을 입력하세요: ')
        # 전화번호 검색
        elif num == 2:
            search = input('전화번호 뒷자리 4자리를 입력하세요: ')
        # 주소 검색
        elif num == 3:
            search = input('주소를 입력하세요: ')
        # 이메일 검색
        elif num == 4:
            search = input('이메일을 입력하세요: ')
        # 그룹 검색
        elif num == 5:
            search = input('그룹을 입력하세요: ')

        # 전역 변수에 저장
        self.num = num
        self.search = search
    
    # 정보 삭제 함수
    def deleteNum(self):
        num = int(input('삭제할 데이터를 입력하세요 (1.이름 2.전화번호 3.주소 4.이메일 5.그룹): '))
        self.inputNum(num)  # 상세 메뉴 함수 호출
        
        for i in range(len(phonelist)):
            # 이름, 이메일, 그룹은 정확히 일치해야 됨
            if num in [1,4,5]:
                if self.search == phonelist[i][num-1]:
                    # 리스트에 존재하면 데이터 삭제
                    del phonelist[i]
            # 전화번호, 주소는 일부만 작성해도 됨
            elif num in [2,3]:
                if self.search in phonelist[i][num-1]:
                    # 리스트에 존재하면 데이터 삭제
                    del phonelist[i]
                print('데이터를 삭제했습니다.')

    # 정보 검색 함수
    def searchNum(self):
        num = int(input('검색할 컬럼을 선택하세요 (1.이름 2.전화번호 3.주소 4.이메일 5.그룹): '))
        self.inputNum(num)  # 상세 메뉴 함수 호출
        
        for i in range(len(phonelist)):
            # 이름, 이메일, 그룹은 정확히 일치해야 됨
            if num in [1,4,5]:
                if self.search == phonelist[i][num-1]:
                    # 리스트에 존재하면 출력
                    print('이름: ' + phonelist[i][0] + ', 전화번호: ' + phonelist[i][1] + ', 주소: ' + phonelist[i][2] + ', 이메일: ' + phonelist[i][3] + ', 그룹: ' + phonelist[i][4])
            # 전화번호, 주소는 일부만 작성해도 됨
            elif num in [2,3]:
                if self.search in phonelist[i][num-1]:
                    print('이름: ' + phonelist[i][0] + ', 전화번호: ' + phonelist[i][1] + ', 주소: ' + phonelist[i][2] + ', 이메일: ' + phonelist[i][3] + ', 그룹: ' + phonelist[i][4])
```

```python
phonelist = [] # 메모리 리스트 생성
info = PhoneInfo() # 객체 생성

try:
    file = open('./phonelist.txt', 'r') # txt파일 열기
    data = file.read()
    # 불필요한 요소 제거 후 str을 list로 변환
    data = data.replace('[','').replace(']','').replace('\n',"','").replace("'","").replace(', ',',').split(',') # str을 list로 변환
    for i in range(0,len(data),5):
        result = list(data[i:i+5]) # 이름, 전화번호, 주소, 이메일, 그룹을 묶어서 리스트화
        phonelist.append(result) # phonelist에 추가

except FileNotFoundError: # Error시 무시하고 진행
    pass

# 메인 함수 반복 실행
while 1:
    try:
        print('******   메뉴   ******\n1. 전화번호 정보 입력\n2. 전화번호 정보 출력\n3. 전화번호 정보 검색\n4. 전화번호 정보 삭제\n5. 전화번호 정보 저장\n6. 프로그램 종료\n**********************')
        select = int(input('메뉴를 선택하세요: '))

        # 각 메뉴 선택시 실행할 함수 호출
        if select == 1:
            info.insertNum()
        elif select == 2:
            info.printNum()
        elif select == 3:
            info.searchNum()
        elif select == 4:
            info.deleteNum()
        elif select == 5:
            info.saveNum()
        elif select == 6:
            print('프로그램을 종료합니다.')
            break
        else: # 다른 키 입력시 리턴
            print('번호를 다시 입력해주세요')
        print()
    except:
        pass # 에러가 발생하더라도 아무 이상없이 진행
```
