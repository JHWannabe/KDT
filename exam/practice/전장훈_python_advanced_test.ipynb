```python
class Phone:
    def __init__(self, name, hp, grp, address):
        self.hp = hp
        self.name = name
        self.grp = grp
        self.address = address
        
    # setter
    def setHp(self, hp):
        self.hp = hp
        
    def setName(self, name):
        self.name = name
        
    def setGrp(self, grp):
        self.grp = grp
        
    def setAddress(self, address):
        self,address = address
        
    # getter
    def getHp(self):
        return self.hp
    
    def getName(self):
        return self.name
    
    def getGrp(self):
        return self.grp
    
    def getAddress(self):
        return self.address
```

```python
import MySQLdb
class Phonebook:
    def __init__(self):     # 생성자
        self.db = None
        
    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'hw')

    def disconnect(self):
        self.db.close()
        
    def insert(self, info):
        self.connect()
        cur = self.db.cursor()
        sql = "insert into phonelist values (%s, %s, %s, %s)"
        data = (info.getName(), info.getHp(), info.getGrp(), info.getAddress())
        cur.execute(sql, data)
        self.db.commit()
        self.disconnect()     
        
    def selectAll(self):
        self.connect()
        cur = self.db.cursor()
        sql = 'select * from phonelist'
        cur.execute(sql)
        row = cur.fetchall()
        self.disconnect()
        return row

    def search(self, num):
        self.connect()
        cur = self.db.cursor()
        sql = 'select * from phonelist where hp like "%' + num +'"'
        cur.execute(sql)
        row = cur.fetchone()
        self.disconnect()
        return row
    
    def update(self, info):
        self.connect()
        cur = self.db.cursor()
        sql = 'update phonelist set hp=%s, grp=%s, address=%s where hp_name=%s'
        data = (info.getHp(), info.getGrp(), info.getAddress(), info.getName())
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 이름이 없습니다')
        self.disconnect()
    
    def delete(self, name):
        self.connect()
        cur = self.db.cursor()
        sql = 'delete from phonelist where hp_name=%s'
        data = (name,)
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('삭제되었습니다')
        else:
            print('해당 이름이 없습니다')
        self.disconnect()
```

```python
class PhoneService:
    def __init__(self):
        self.dao = Phonebook()
    
    def insertHp(self):
        name = input('이름을 입력하세요: ')
        hp = input('핸드폰 번호를 입력하세요: ')
        grp = input('그룹을 입력하세요: ')
        address = input('주소를 입력하세요: ')
        info = Phone(name, hp, grp, address)
        self.dao.insert(info)
    
    def printAll(self):
        datas = self.dao.selectAll()
        print(datas)
    
    def searchHp(self):
        num = input('조회할 번호 뒷자리 4자리를 입력하세요: ')
        hp = self.dao.search(num)
        if hp:
            print(hp)
        else:
            print('찾는 번호가 없습니다')
    
    def editHp(self):
        name = input('수정할 이름을 입력하세요: ')
        row = self.dao.search(name)
        if row == None:
            print('수정할 단어를 찾지 못했습니다')
        else:
            col = {1:"hp", 2:"grp", 3:"address"}
            select = int(input('수정할 정보를 입력하세요(1.전화번호 2.그룹 3.주소): '))
            new_data = input('새로운 정보를 입력하세요: ')
            self.dao.update(num, col[select], new_data)
    
    def delHp(self):
        name = input('삭제할 이름을 입력하세요: ')
        self.dao.delete(name)
```

```python
class Menu:
    def __init__(self):
        self.service = PhoneService()
    
    def run(self):
        while True:
            try:
                menu = int(input('1.등록하기 2.출력하기 3.검색하기 4.수정하기 5.삭제하기 6.종료하기'))
                if menu == 1:
                    self.service.insertHp()
                elif menu == 2:
                    self.service.printAll()
                elif menu == 3:
                    self.service.searchHp()
                elif menu == 4:
                    self.service.editHp()
                elif menu == 5:
                    self.service.delHp()
                elif menu == 6:
                    break
            except Exception as e:
                print(e)
                print('다시 입력하세요')
```
