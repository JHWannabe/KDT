# 학생 DTO

```python
class Student:
    def __init__(self, s_id, s_name, s_hp, s_email, s_address):
        self.id = s_id # 학번
        self.name = s_name # 학생 이름
        self.hp = s_hp # 학생 연락처
        self.email = s_email # 학생 이메일
        self.address = s_address # 학생 주소

    # setter
    def setSid(self, id):
        self.id = id

    def setSname(self, name):
        self.name = name

    def setShp(self, hp):
        self.hp = hp

    def setSemail(self, email):
        self.email = email

    def setSaddress(self, address):
        self.address = address

    # getter
    def getSid(self):
        return self.id

    def getSname(self):
        return self.name

    def getShp(self):
        return self.hp

    def getSemail(self):
        return self.email

    def getSaddress(self):
        return self.address
```

# 성적 DTO

```python
class Grade:
    def __init__(self, g_id, java, python, c, total, average):
        self.id = g_id # 학생 학번
        self.java = java # 자바 점수
        self.python = python # 파이썬 점수
        self.c = c # C언어 점수
        self.total = total # 총점
        self.average = average # 평균

    # setter
    def setGid(self, id):
        self.id = id

    def setJava(self, java):
        self.java = java

    def setPython(self, python):
        self.python = python

    def setC(self, c):
        self.c = c

    def setTotal(self, total):
        self.total = total

    def setAverage(self, average):
        self.average = average

    # getter
    def getGid(self):
        return self.id

    def getJava(self):
        return self.java

    def getPython(self):
        return self.python

    def getC(self):
        return self.c

    def getTotal(self):
        return self.total

    def getAverage(self):
        return self.average
```

# 학생 DAO

```python
import MySQLdb


class StudentList:
    def __init__(self):
        self.db = None

    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'hw') # MySQL 접속

    def disconnect(self):
        self.db.close()

    # SQL 삽입
    def insert(self, st):
        self.connect()
        cur = self.db.cursor()
        sql = "insert into Student (학번, 이름, 연락처, 이메일, 주소) values (%s, %s, %s, %s, %s)"
        data = (st.getSid(), st.getSname(), st.getShp(), st.getSemail(), st.getSaddress())
        cur.execute(sql, data)
        self.db.commit()
        self.disconnect()

    # SQL 모두 출력
    def printAll(self):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from vw_student order by 평균 desc, 학번'
        cur.execute(sql)
        count = 0 # 테이블에 등록된 학생 수 집계 변수
        while 1:
            st = cur.fetchone()
            if st:
                if st['평균'] == None:
                    print(f"학번: {st['학번']}, 이름: {st['이름']}, 연락처: {st['연락처']}, 이메일: {st['이메일']}, 주소: {st['주소']}")
                else:
                    print(f"학번: {st['학번']}, 이름: {st['이름']}, 연락처: {st['연락처']}, 이메일: {st['이메일']}, 주소: {st['주소']}, 평균: {st['평균']}점")
                count += 1
            else:
                break
        self.disconnect()
        return count

    def delete(self, num):
        self.connect()
        cur = self.db.cursor()
        sql = 'delete from Grade where 학번=%s'
        result = cur.execute(sql, (num,))
        sql = 'delete from Student where 학번=%s'
        result = cur.execute(sql, (num,))
        self.db.commit()
        if result > 0:
            print('삭제되었습니다')
        else:
            print('해당 정보가 없습니다')
        self.disconnect()

    def edit(self, num, col, new_data):
        self.connect()
        cur = self.db.cursor()
        sql = 'update Student set ' + col + '=%s where 학번=%s'
        data = (new_data, num)
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 정보가 없습니다')
        self.disconnect()

    def search(self, num):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from vw_student where 학번=%s'
        cur.execute(sql, (num,))
        row = cur.fetchone()
        self.disconnect()
        return row
```

# 성적 DAO

```python
class GradeList:
    def __init__(self):
        self.db = None

    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'hw')

    def disconnect(self):
        self.db.close()

    def insert(self, gr):
        self.connect()
        cur = self.db.cursor()
        sql = "insert into Grade (학번, 자바, 파이썬, C언어, 총점, 평균) values (%s, %s, %s, %s, %s, %s)"
        data = (gr.getGid(), gr.getJava(), gr.getPython(),
                gr.getC(), gr.getTotal(), gr.getAverage())
        cur.execute(sql, data)
        self.db.commit()
        self.disconnect()

    def printAll(self):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from vw_grade'
        cur.execute(sql)
        while 1:
            gr = cur.fetchone()
            if gr:
                print(
                    f"학번: {gr['학번']}, 자바: {gr['자바']}점, 파이썬: {gr['파이썬']}점, C언어: {gr['C언어']}점, 총점: {gr['총점']}점, 평균: {gr['평균']}점")
            else:
                break
        self.disconnect()

    def delete(self, num):
        self.connect()
        cur = self.db.cursor()
        sql = 'delete from Grade where 학번=%s'
        result = cur.execute(sql, (num,))
        self.db.commit()
        if result > 0:
            print('삭제되었습니다')
        else:
            print('해당 정보가 없습니다')
        self.disconnect()

    def edit(self, num, java, python, c):
        self.connect()
        cur = self.db.cursor()
        total = int(java) + int(python) + int(c)
        average = total / 3
        sql = 'update Grade set 자바=%s, 파이썬=%s, C언어=%s, 총점=%s, 평균=%s where 학번=%s'
        data = (java, python, c, total, round(average,1), num)
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 정보가 없습니다')
        self.disconnect()

    def search(self, num):
        self.connect()
        cur = self.db.cursor(MySQLdb.cursors.DictCursor)
        sql = 'select * from Grade where 학번=%s'
        cur.execute(sql, (num,))
        row = cur.fetchone()
        self.disconnect()
        return row
```

# 학생 Service

```python
class StService:
    def __init__(self):
        self.st = StudentList()  # 학생 리스트

    # 학생 정보 입력
    def insertStudent(self):
        id = input('학번을 입력하세요(필수): ')
        name = input('이름을 입력하세요(필수): ')
        hp = input('전화번호을 입력하세요(선택): ')
        if hp == '':
            hp = '없음'
        email = input('이메일을 입력하세요(선택): ')
        if email == '':
            email = '없음'
        address = input('주소를 입력하세요(선택): ')
        if address == '':
            address = '없음'

        new = Student(id, name, hp, email, address)
        self.st.insert(new)

    # 학생 정보 출력
    def printAll(self):
        count =self.st.printAll()
        print('현재 등록된 학생 수: %d명' % count)

    # 학생 정보 수정
    def editStudent(self):
        num = input('정보를 수정할 학생의 학번을 입력하세요: ')
        row = self.st.search(num)
        if row == None:
            print('수정할 영화가 없습니다.')
        else:
            col = {1: "이름", 2: "연락처", 3: "이메일", 4: "주소"}
            select = int(input('수정할 정보를 입력하세요(1.이름 2.연락처 3.이메일 4.주소): '))
            new_data = input('새로운 정보를 입력하세요: ')
            self.st.edit(num, col[select], new_data)

    # 학생 정보 삭제
    def delStudent(self):
        num = input('정보를 삭제할 학생의 학번을 입력하세요: ')
        self.st.delete(num)

    # 학생 정보 검색
    def searchStudent(self):
        num = input('정보를 검색할 학생의 학번을 입력하세요: ')
        st = self.st.search(num)
        if st['평균'] == None:
            print(f"학번: {st['학번']}, 이름: {st['이름']}, 연락처: {st['연락처']}, 이메일: {st['이메일']}, 주소: {st['주소']}")
        else:
            print(f"학번: {st['학번']}, 이름: {st['이름']}, 연락처: {st['연락처']}, 이메일: {st['이메일']}, 주소: {st['주소']}, 평균: {st['평균']}점")
```

# 성적 Service

```python
class GrService:
    def __init__(self):
        self.gr = GradeList()  # 성적 리스트

    # 성적 입력
    def insertGrade(self):
        id = input('학번을 입력하세요: ')
        java = int(input('Java 점수를 입력하세요: '))
        python = int(input('Python 점수를 입력하세요: '))
        c = int(input('C언어 점수를 입력하세요: '))
        total = java + python + c
        average = total / 3
        new = Grade(id, java, python, c, total, round(average,1)) # 평균은 소수점 1자리까지만 저장
        self.gr.insert(new)

    # 성적 출력
    def printAll(self):
        self.gr.printAll()

    # 성적 수정
    def editGrade(self):
        num = input('성적을 수정할 학생의 학번을 입력하세요: ')
        row = self.gr.search(num)
        if row == None:
            print('수정할 학생이 없습니다.')
        else:
            java = input('자바 점수를 입력하세요: ')
            python = input('파이썬 점수를 입력하세요: ')
            c = input('C언어 점수를 입력하세요: ')
            self.gr.edit(num, java, python, c)

    # 성적 삭제
    def delGrade(self):
        num = input('성적를 삭제할 학생의 학번을 입력하세요: ')
        self.gr.delete(num)

    # 성적 검색
    def searchGrade(self):
        num = input('성적를 검색할 학생의 학번을 입력하세요: ')
        gr = self.gr.search(num)
        print(f"학번: {gr['학번']}, 자바: {gr['자바']}점, 파이썬: {gr['파이썬']}점, C언어: {gr['C언어']}점, 총점: {gr['총점']}점, 평균: {gr['평균']}점")
```

# Main 클래스

```python
class Menu:
    def __init__(self):
        self.st = StService()  # 학생관리
        self.gr = GrService()  # 성적관리

    def run(self):
        while 1:
            try:
                print('*********** 메뉴 ***********\n1. 학생 관리\n2. 성적 관리\n0. 종료')
                menu1 = input('>>> ')
                print()

                # 학생 관리 메뉴
                if menu1 == '1':
                    while 1:
                        print(
                            '1. 학생정보 입력\n2. 학생정보 출력\n3. 학생정보 수정\n4. 학생정보 삭제\n5. 학생정보 검색\n0. 뒤로가기')
                        menu2 = input('>>> ')
                        # 입력
                        if menu2 == '1':
                            self.st.insertStudent()
                        # 출력
                        elif menu2 == '2':
                            self.st.printAll()
                        # 수정
                        elif menu2 == '3':
                            self.st.editStudent()
                        # 삭제
                        elif menu2 == '4':
                            self.st.delStudent()
                        # 검색
                        elif menu2 == '5':
                            self.st.searchStudent()
                        # 뒤로가기
                        elif menu2 == '0':
                            break
                        # 다른 키 입력시 에러
                        else:
                            print('다시 입력하세요')

                        print()

                # 성적 관리 메뉴
                elif menu1 == '2':
                    while 1:
                        print(
                            '1. 성적 입력\n2. 성적 출력\n3. 성적 수정\n4. 성적 삭제\n5. 성적 검색\n0. 뒤로가기')
                        menu2 = input('>>> ')
                        # 입력
                        if menu2 == '1':
                            self.gr.insertGrade()
                        # 출력
                        elif menu2 == '2':
                            self.gr.printAll()
                        # 수정
                        elif menu2 == '3':
                            self.gr.editGrade()
                        # 삭제
                        elif menu2 == '4':
                            self.gr.delGrade()
                        # 검색
                        elif menu2 == '5':
                            self.gr.searchGrade()
                        # 뒤로가기
                        elif menu2 == '0':
                            break
                        # 다른 키 입력시 에러
                        else:
                            print('다시 입력하세요')

                        print()

                # 프로그램 종료
                elif menu1 == '0':
                    break
                # 다른 키 입력시 에러
                else:
                    print('다시 입력하세요')

                print()

            except:
                print('에러가 발생했습니다. 다시 시도해주세요.\n')
```

```python
model = Menu()
model.run()
```
